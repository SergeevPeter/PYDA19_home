### Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
df = pd.read_csv(r'c:\\Users\\Petr\\Desktop\\Python\\Курс - Python (от Нетология)\\PYDA_19\\02 NumPy, pandas, ML\\03-Функции и работа с данными\\ml-latest-small\\ratings.csv',sep = ',')
df.head()
```

```python
def fn_segment(arg1):
    if arg1 <= 2:
        return "low"
    elif arg1 <= 4:
        return "medium"
    elif arg1 > 4:
        return "high"
```

```python
df['class'] = df['rating'].apply(fn_segment)
df['class'].value_counts()
```

### Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.

```python
df2 = pd.read_csv(r'c:\\Users\\Petr\\Desktop\\Python\\Курс - Python (от Нетология)\\PYDA_19\\02 NumPy, pandas, ML\\03-Функции и работа с данными\\ml-latest-small\\keywords.csv',sep = ',')
```

```python
df2.head(6)
```

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

def fn_1_(arg1):
    for i1,i2 in geo_data.items():
        for k1 in arg1:
            if k1 in i2:
                return i1
            else:
                return 'Undefined'

df2['region'] = df2['keyword'].apply(lambda word_: word_.lower().split(' ')).apply(fn_1_)
df2.sort_values(['region'], ascending = False)
df2['region'].value_counts()
```

### Задание 3 (бонусное)

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл movies.csv и ratings.csv из базы https://grouplens.org/datasets/movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

1.В переменную years запишите список из всех годов с 1950 по 2010.
2.Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
3.для каждой строки пройдите по всем годам списка years
--если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
--если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
--Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

4.Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга.

```python
r_ = pd.read_csv(r'c:\\Users\\Petr\\Desktop\\Python\\Курс - Python (от Нетология)\\PYDA_19\\02 NumPy, pandas, ML\\02-Библиотека pandas\\ml-latest-small\\ratings.csv')
m_ = pd.read_csv(r'c:\\Users\\Petr\\Desktop\\Python\\Курс - Python (от Нетология)\\PYDA_19\\02 NumPy, pandas, ML\\02-Библиотека pandas\\ml-latest-small\\movies.csv')
```

```python
ds = r_.merge(m_,on='movieId',how='left')
ds.head(2)
```

```python
years_ = []
for x in range(1950,2011):
    years_.append(x) 
```

```python
ds['year_step1'] = ds['title'].apply(lambda word: word.replace(")","").replace("(","").split(' '))
ds.head(2)
```

```python
def fn_production_year(arg1):
    for i1 in years_:
        if str(i1) in arg1:
            return i1
    return 1900
```

```python
ds['year_step2'] = ds['year_step1'].apply(fn_production_year)
ds.head(7)
```

```python
ds_mean = ds.groupby('year_step2').agg(['mean'])['rating'].reset_index().\
        sort_values('mean', ascending=False)

ds_mean.head(7)
```
